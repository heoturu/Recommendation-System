### Similarity (유사도)
- Euclidean Distance Similarity
- Cosine Similarity
- Compare Euclidean Distance & Cosine Similarity

```python
import numpy as np
import pandas as pd
from scipy import spatial
from PIL import Image as pimg
```

```python
# Sample Data
vector_1 = np.array([1, 2, 3, 4, 5])
vector_2 = np.array([5, 4, 3, 2, 1])
vector_3 = np.array([2, 3, 4, 5, 6])
vector_4 = np.array([11, 19, 28, 32, 47])
```

#### 1. Euclidean Distance Similarity
- 벡터 사이의 거리
- 거리가 작으면 가까운 위치에 있기 때문에 비슷한 특징을 가짐

```python
pimg.open('Euclidean Distance Similarity.png')
```

```python
# 계산과정
print((vector_1 - vector_2))
print((vector_1 - vector_2) ** 2)
print(sum((vector_1 - vector_2) ** 2))
print('=' * 20)
print(np.sqrt(sum((vector_1 - vector_2) ** 2)))
```

```python
# Numpy 활용
np.linalg.norm(vector_1 - vector_2)
```

```python
def eculidean_similarity(vector_1, vector_2):
    return np.linalg.norm(vector_1 - vector_2)
```

```python
eculidean_similarity(vector_1, vector_2)
```

#### 2. Cosine Similarity
- 벡터 사이의 cosine 각
- 각이 작으면 같은 방향성을 가지기 때문에 비슷한 특징을 가짐
- 벡터의 내적 / 벡터의 크기

```python
pimg.open('Cosine Similarity.png')
```

```python
# numerator - Vector Innor Product
num = np.dot(vector_1, vector_2)    # sum(vector_1 * vector_2)

# denominator
denom = np.sqrt(sum(vector_1 ** 2)) * np.sqrt(sum(vector_2 ** 2))

cosine_sim = num / denom
cosine_sim
```

```python
# scipy 이용
1 - spatial.distance.cosine(vector_1, vector_2),\
1 - spatial.distance.cosine(vector_1, vector_3),\
1 - spatial.distance.cosine(vector_1, vector_4)
```

코사인 유사도에서는 벡터1과 벡터4가 가장 유사 (세타 활용)
